```python
import pandas as pd
import numpy as math
import matplotlib.pyplot as plt

obesity = pd.read_csv("../data/obesity.csv")

obesity
# find zip codes with no data missing, remove ages under 18
no_miss_zips = obesity[obesity.age !=
                       "05-17"].dropna().groupby("zip").size().reset_index(name='n')

no_miss_zips = no_miss_zips[no_miss_zips["n"] == 8]


full_data_obesity = obesity[obesity.age !="05-17"].dropna()

full_data_obesity = full_data_obesity[obesity["zip"].isin(no_miss_zips.zip)]

full_data_obesity = full_data_obesity[full_data_obesity["pop"] > 5000]

full_data_obesity = full_data_obesity[full_data_obesity["sex"] == "male"].assign(
    pct_obese=full_data_obesity["obese"] / full_data_obesity["pop"])

full_data_obesity = full_data_obesity[full_data_obesity["pct_obese"] > 0.10]

full_data_obesity["zip"] = full_data_obesity["zip"].astype(str)

print(full_data_obesity)
```

```python
plt.bar(x=full_data_obesity["zip"],
        height=full_data_obesity["pct_obese"], width=0.6, color="maroon")
plt.title("Male Obesity Rate by Zipcode (>10%)")
plt.xlabel("Zipcode")
plt.ylabel("Obesity Rate")
plt.show()
```
